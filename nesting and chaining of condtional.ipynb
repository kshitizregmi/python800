# nesting of conditionals

```python
num = 0

if num < 0:
    print("negative")
    
else:
    if num == 0:
        print("zero")
        
    else:
        print("Positive")
    
```

# chaining of conditionals: if... elif .... else

```python
num = 5

if num < 0:
    print("negative")

elif num == 0:
    print("Zero")
    
else:
    print("Positive")
```
